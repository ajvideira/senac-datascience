```python
import pandas as pd

df_clientes = pd.read_csv('../input/clientes.csv')
df_pedidos = pd.read_csv('../input/pedidos.csv')

df_merge = df_pedidos.merge(df_clientes, 'left', 'id_cliente')

display(df_merge)

valor_por_estado = df_merge.groupby('estado').agg(total_vendido=('valor', 'sum')).reset_index()

display(valor_por_estado)

print(f'Estado com mais venda: {valor_por_estado['total_vendido'].idxmax()}, total de R$ {valor_por_estado['total_vendido'].max()}')

vendas_por_estado = df_merge.groupby('estado').agg(total_vendido=('id_pedido', 'count')).reset_index()

display(vendas_por_estado)
print(f'Estado com mais pedidos: {vendas_por_estado['total_vendido'].idxmax()}, total de {vendas_por_estado['total_vendido'].max()} pedidos')
```
