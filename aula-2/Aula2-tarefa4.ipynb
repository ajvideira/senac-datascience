```python
import pandas as pd

df_funcionarios = pd.read_csv('../input/funcionarios.csv')
display(df_funcionarios)

df_funcionarios.isnull().sum()

df_funcionarios['salario'] = df_funcionarios['salario'].fillna(df_funcionarios['salario'].mean()).round()
display(df_funcionarios)

df_funcionarios['departamento'] = df_funcionarios['departamento'].fillna('Não Alocado')
display(df_funcionarios)

print('Linhas duplicadas: ',df_funcionarios.duplicated().sum())

df_funcionarios = df_funcionarios.drop_duplicates()
display(df_funcionarios)
```
