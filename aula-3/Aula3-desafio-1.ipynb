```python
import pandas as pd

df_pedidos = pd.read_json('../input/pedidos_simples.json')

display(df_pedidos)

df_clientes = pd.read_json('../input/clientes_simples.json')

display(df_clientes)

df_completo = df_pedidos.merge(df_clientes, 'left', 'id_cliente')

display(df_completo)

df_por_cidade = df_completo.groupby('cidade').agg(valor_total=('valor', 'sum'), qtd_pedidos=('id_pedido', 'count')).reset_index()

display(df_por_cidade)

df_por_cidade.set_index('cidade').to_json('report_por_cidade.json', orient='index', indent=4)
```
