```python
import pandas as pd
import json

with open('../input/dados_pedidos.json', 'r', encoding="utf-8") as json_file:
    data = json.load(json_file)

df_pedidos = pd.json_normalize(data, record_path=['detalhes'], meta=['pedido_id', 'data_pedido', 'status', ['cliente', 'nome']], sep='.')

assert len(df_pedidos)>0, 'ERRO: dataframe vazio!'

df_pedidos['preco_unitario'] = pd.to_numeric(df_pedidos['preco_unitario'], errors='coerce')
df_pedidos['data_pedido'] = pd.to_datetime(df_pedidos['data_pedido'], errors='coerce')
df_pedidos['desconto_aplicado'] = pd.to_numeric(df_pedidos['desconto_aplicado'], errors='coerce').fillna(0)

df_pedidos['total_item_bruto'] = df_pedidos['qtd'] * df_pedidos['preco_unitario']
df_pedidos['total_item_liquido'] = df_pedidos['total_item_bruto'] * df_pedidos['desconto_aplicado']

df_pedidos.drop(columns=['desconto_aplicado'], inplace=True)

display(df_pedidos)

df_pedidos.info()
```
