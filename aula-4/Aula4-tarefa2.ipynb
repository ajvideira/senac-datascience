```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def clean_number(raw):
    return raw.replace('$', '').replace(' ','').replace(',','').replace('\n', '')

proxies = {
    'http': 'http://localhost:3128',
    'https': 'http://localhost:3128',
}

headers = {
 'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'

response = requests.get(url, proxies=proxies, headers=headers)

content = response.text

soup = BeautifulSoup(content, 'lxml')

trs = soup.select_one('.wikitable').select('tr')[2:]

companies = []

industry = None
revenue = 0

for tr in trs:
    if tr.select('td')[1].get('rowspan') or tr.select('td')[1].select_one('a'):
        industry = tr.select('td > a')[1].text
        revenue = clean_number(tr.select('td')[2].text)
    else:
        revenue = clean_number(tr.select('td')[1].text)

    company = {
        'rank': int(tr.select('th')[0].text.replace('\n', '')),
        'name': tr.select('td > a')[0].text,
        'industry': industry,
        'revenue': float(revenue)
    }

    companies.append(company)

df_companies = pd.DataFrame(companies)

df_companies.to_csv('companies.csv', index=False)

display(df_companies)
```
